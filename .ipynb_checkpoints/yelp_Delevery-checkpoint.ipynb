```python
# call resturants that do delevery:

url = "https://api.yelp.com/v3/transactions/delivery/search"
API = api_key
params = {"location":"cleveland"}
headers = {"Authorization":"bearer %s" % API} 

# call the request
response = requests.get(url=url, params=params, headers=headers).json()

pprint(response, indent = 4)

pprint(response) 
```

```python
# convert json out come to DF

response =response['businesses']
DF = pd.DataFrame.from_dict(response)
DF.head(2)
```

```python
DF.columns
```

```python
# split the coordinates col into lat and lng.

lats = []
lngs = []

coords = DF["coordinates"]
coords
```

```python
# loop in the coordinates col to split it.

for loc in coords:
     
    lat = loc["latitude"]
    lng = loc["longitude"]
    print (lat , lng)
    lats.append(lat)
    lngs.append(lng)
    
latitude = lats
longitude = lngs

print (latitude , longitude)
```

```python
latitude = pd.DataFrame(latitude)
latitude
```
