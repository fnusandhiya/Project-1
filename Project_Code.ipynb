```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import random
from pprint import pprint
from yelpapi import YelpAPI
from gql import gql, Client


# Yelp API KEY
from API import api_key
api_key
```

```python
# https://www.yelp.com/developers/documentation/v3/business_search -- Link for API search

url = 'https://api.yelp.com/v3/businesses/search'
params = {'location': 'Cleveland',
        'term': 'restaurants'}
API = api_key
headers = {'Authorization': 'bearer %s' % API}
response = requests.get(url=url, params=params, headers=headers).json()
response=response['businesses']
df=pd.DataFrame.from_dict(response)
df.head()
```

```python
coord_df = pd.DataFrame(columns=['latitude','longitude'])

for (i,r) in df.iterrows():
    e = r['coordinates']
    coord_df.loc[i] = [e['latitude'], e['longitude']]
df = pd.concat([df, coord_df], axis=1)
df.head()
```

```python

df = df[['alias', 'categories', 'display_phone', 'distance', 'id','image_url', 
         'is_closed','location', 'name', 'phone', 'price', 'rating', 'review_count',
         'transactions','url','latitude','longitude']]
df.head()
```

```python
zip_df = pd.DataFrame(columns=['zip_code'])

for (i,r) in df.iterrows():
    e = r['location']
    zip_df.loc[i] = [e['zip_code']]
df = pd.concat([df, zip_df], axis=1)
df.head()
```
